# 数据清洗

```python
import pandas as pd
```

```python
# 读取 pokemon151_v3.csv 数据集
pokemon_path = "../data/raw/pokemon151_v3.csv"
pokemon_df = pd.read_csv(pokemon_path, encoding="utf-8")
pokemon_df.head()
```

```python
# 读取 move_v2.csv 数据集
move_path = "../data/raw/move_v2.csv"
move_df = pd.read_csv(move_path, encoding="utf-8")
move_df.head()
```

```python
# 读取 learn_v2.csv 数据集
learn_path = "../data/raw/learn_v2.csv"
learn_df = pd.read_csv(learn_path, encoding="utf-8")
learn_df.head()
```

## 处理 move 表

```python
# 将 move_df 中的 category 列中的数值替换为对应的字符串
move_df.loc[move_df["category"] == '1', "category"] = "physical"
move_df.loc[move_df["category"] == '2', "category"] = "special"
move_df.loc[move_df["category"] == '3', "category"] = "status"
move_df.head()
```

## 处理 learn 表

### 检查 learn.csv 中的重复行

```python
# 以pokemon_id, move_id, method为依据，检查重复行总数
cols_v1 = ["pokemon_id", "move_id", "method"]

dup_rows_v1 = learn_df[learn_df.duplicated(subset=cols_v1, keep=False)].sort_values(cols_v1)
print(f"重复行总数: {dup_rows_v1.shape[0]}")
dup_rows_v1
```

```python
before = len(learn_df)
learn_df_v1 = learn_df.drop_duplicates(subset=cols_v1, keep="first").reset_index(drop=True)
after = len(learn_df_v1)
print(f"保留第一条 — 删除: {before - after} 行 （前: {before}, 后: {after}）")
learn_df_v1.head()
```

### 检查发现 method_transfer 列意义不大且容易出问题，删去此列

```python
# 显示 method_transfer 列值不为 NAN 的所有行
filtered_rows = learn_df_v1[learn_df_v1["method_transfer"].notnull()]
print(f"method_transfer 不为 NAN 的行数: {filtered_rows.shape[0]}")
filtered_rows
```

```python
learn_df_v2 = learn_df_v1.drop(columns=["method_transfer"])
learn_df_v2.head()
```

### 更新 learn_id

```python
learn_df_v2["learn_id"] = range(1, len(learn_df_v2) + 1)
learn_df_v2
```

```python
value = 1
pokemon_learn = learn_df_v2[learn_df_v2["pokemon_id"] == value]
print(f"id 为 {value} 的宝可梦的技能学习情况: {pokemon_learn.shape[0]}")
pokemon_learn
```

## 处理 pokemon 表

### 更新 latest_generation

```python
# 计算每个 pokemon_id 的最大 gen
max_gen = learn_df.groupby("pokemon_id", as_index=False)["gen"].max()
max_gen.rename(columns={"gen": "latest_generation"}, inplace=True)

# 赋值到 pokemon_df
pokemon_df["latest_generation"] = pokemon_df["pokemon_id"].map(
    learn_df.groupby("pokemon_id")["gen"].max()
)
pokemon_df
```

## 导出清洗后的数据

```python
data_path = "../data/processed/"
pokemon_df.to_csv(data_path + 'pokemon.csv', index=False, encoding='utf-8')
move_df.to_csv(data_path + 'move.csv', index=False, encoding='utf-8')
learn_df_v2.to_csv(data_path + 'learn.csv', index=False, encoding='utf-8')
```
